```python
from sklearn.datasets import make_blobs
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
#gera numeros aleatorios, 200 numeros aleatorios com 5 clusters
x_random, y_random = make_blobs(n_samples = 200, centers = 5)
x_random
```

```python
kmeans_blobs = KMeans(n_clusters = 5)
kmeans_blobs.fit(x_random)
```

```python
rotulos = kmeans_blobs.predict(x_random)
rotulos
```

```python
centroides = kmeans_blobs.cluster_centers_
centroides
```

```python
grafico_pontos = px.scatter(x = x_random[:,0], y = x_random[:,1], color = rotulos)
#grafico_pontos.show()
grafico_centroides = px.scatter(x = centroides[:,0], y = centroides[:,1], size = [3,3,3,3,3])
#grafico_centroides.show()
grafico_final = go.Figure(data = grafico_pontos.data + grafico_centroides.data)
grafico_final.show()
```
